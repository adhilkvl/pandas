# handeling missing values

```python
data={
    "no":[1,2,None,4,5,6,7,8,9,10,11],
    "name":["shaji","lalu",None,"vani","meera","bilu","ram","rose","teena","anusha","shabna"],
    "age":[3,6,None,9,None,7,5,41,12,None,28]
}
```

```python
df=pd.DataFrame(data)
print(df)
```

```python
df.isnull()
```

```python
df.isnull().sum()
```

```python
df.isnull().sum().sum()
```

## fillna ()

```python
result=df.fillna(value=0)
print(result)
```

### fill with prv value

```python
result=df.fillna(method='ffill')
print(result)
```

### fill with after value

```python
result=df.fillna(method='bfill')
print(result)
```

```python
result=df.fillna(method='ffill',axis=1)
print(result)
```

# dropna ()

#### drop all the missing values or none values in a row

```python
result=df.dropna(how='all')
print(result)
```

#### drop any the missing values or none values in a row

```python
result=df.dropna(how='any',ignore_index=True)
print(result)
```

### threshod method

```python
result=df.dropna(thresh=2)
print(result)
```

```python
result=df.dropna(thresh=3)
print(result)
```

```python
result=df.dropna(thresh=9,axis=1)
print(result)
```

## subset method

```python
result=df.dropna(subset=['name'])
print(result)
```

```python
result=df.dropna(subset=['age'])
print(result)
```
